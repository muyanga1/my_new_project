```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
SEEDS_DATA_URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00236/seeds_dataset.txt'
seeds_df = pd.read_csv(SEEDS_DATA_URL, sep="\s+", header=None)
column_names = ['f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7', 'target']
seeds_df.columns = column_names
#seeds_df.shape
seeds_df
```

```python
# Generating the array output for explicitly selected predictors and writing the arra
from sklearn.linear_model import LogisticRegression
```

```python
X = seeds_df[['f1']]
y = seeds_df['target']
             
lr = LogisticRegression()
lr.fit(X, y)
preds = lr.predict(X) # will output array with integer values.
#display(preds)
display(preds.shape)
import csv
wtr = csv.writer(open('out_file3.csv', 'w'), delimiter=',', lineterminator='\n')
for x in preds :
    wtr.writerow ([x])
```

```python
# Same as above. But here we are making the list comprehension to scan through the f
from sklearn.linear_model import LogisticRegression
import csv
features = ('f1','f2','f3','f4','f5','f6','f7','target')
#features = ('Area','Perimeter','Compactness','Length','Width','Asymmetry Coefft','Kern
new_prediction = np.zeros(210)
for i in range(7):
    print('\nthe new value of i is :',i)
    print( 'before the iteration, the feature is:',(features[i]) )
    seeds_df[(features[i])].hist(bins=100)
    X = seeds_df[[(features[i])]]
    y = seeds_df['target']
    lr = LogisticRegression()
    lr.fit(X, y)
    preds = lr.predict(X) # will output array with integer values.
    new_prediction = np.vstack([new_prediction,preds])
    print('after the iteration i =',i)
    display(new_prediction)
    np.savetxt('output.csv', new_prediction.transpose(), delimiter=',', fmt='%d')
```
